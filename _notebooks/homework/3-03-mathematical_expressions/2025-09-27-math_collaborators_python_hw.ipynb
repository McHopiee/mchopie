  <pre><code>def compute_over_sum(x, y):
    if x + y == 0:
        return None  # undefined if denominator is 0
    return (x**2 + y**2) / (x + y)
</code></pre>

  <p><b>2) PEMDAS for <code>result = 6 + 4 * 2 % 5</code></b></p>
  <ol>
    <li>4 * 2 ‚Üí 8</li>
    <li>8 % 5 ‚Üí 3 (remainder of 8 / 5)</li>
    <li>6 + 3 ‚Üí 9</li>
  </ol>
  <p><b>Final Answer:</b> 9</p>

  <p><b>3) Function: calculate_mod_sum(a, b, c)</b></p>
  <pre><code>def calculate_mod_sum(a, b, c):
    if b == 0 or c == 0:
        return None  # cannot modulo by zero
    return (a % b) + (b % c)
</code></pre>

  <pre><code>def fib_like(n):
    if n == 1:
        return 3
    if n == 2:
        return 5
    a, b = 3, 5
    for _ in range(3, n + 1):
        a, b = b, a + b
    return b
</code></pre>
