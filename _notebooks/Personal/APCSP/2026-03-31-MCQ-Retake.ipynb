## 1. Game Piece / Counter Problem
A game piece moves left or right depending on the color of the space:

- Yellow -> move 3 left
- Black -> move 1 left
- Green -> move 2 right

A counter increases each time the piece moves. The game starts on the rightmost black space and ends when the piece lands on red or leaves the board.

<div style="margin: 1rem 0 1.1rem 0; padding: 0.95rem 1rem; border-radius: 16px; background: linear-gradient(135deg, #ecfeff 0%, #f0fdfa 100%); border: 1px solid #99f6e4;">
  <div style="font-size: 0.78rem; text-transform: uppercase; letter-spacing: 0.08em; color: #0f766e; margin-bottom: 0.35rem;">Question</div>
  <strong>What is the final value of the counter?</strong>
</div>

- Chosen answer: B (3)
- Correct answer: 4
- Why this is wrong: The counter increments after every move, not just certain moves.
- Correct reasoning:
  - Black -> move 1 left -> counter = 1
  - Yellow -> move 3 left -> counter = 2
  - Green -> move 2 right -> counter = 3
  - Green -> move 2 right -> lands on red -> counter = 4

<div style="margin: 1rem 0 0.9rem 0; height: 1px; background: linear-gradient(90deg, #cbd5e1 0%, transparent 100%);"></div>

- Core mistake: The full sequence of moves was not completed.

## 3. MOD Code Problem
A list is processed with:

result <- result + (item MOD 2)

<div style="margin: 1rem 0 1.1rem 0; padding: 0.95rem 1rem; border-radius: 16px; background: linear-gradient(135deg, #f5f3ff 0%, #f8fafc 100%); border: 1px solid #c4b5fd;">
  <div style="font-size: 0.78rem; text-transform: uppercase; letter-spacing: 0.08em; color: #7c3aed; margin-bottom: 0.35rem;">Question</div>
  <strong>What value is displayed?</strong>
</div>

- Chosen answer: B (4)
- Correct answer: 3
- Why this is wrong: item MOD 2 does not count even numbers.
- Correct reasoning:
  - Even -> MOD 2 = 0
  - Odd -> MOD 2 = 1
- So the program counts odd numbers.
- List: [4, 2, 5, 4, 2, 3, 1]
- Odd values: 5, 3, 1 -> total = 3

<div style="margin: 1rem 0 0.9rem 0; height: 1px; background: linear-gradient(90deg, #cbd5e1 0%, transparent 100%);"></div>

- Core mistake: Misinterpreting MOD as counting evens.
