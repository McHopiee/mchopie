# MCQ Wrap-Up Expectations and Plan

## What I need to study more (based on what I missed)
From my wrong answers, I can see clear patterns in the topics I need to strengthen.

1. Boolean logic and logic gates
- I need more practice reading multi-step gate circuits and checking when final outputs are true/false.

2. Algorithm tracing and loop flow
- I need to improve at tracing variable updates line by line, especially in REPEAT/UNTIL and IF blocks.

3. Conditionals and range checks
- I need to review compound Boolean expressions (like checking whether a value is within a range).

4. Binary, bits, and data representation
- I need stronger fluency converting decimal/binary quickly (RGB values, bit-length capacity, metadata vs data).

5. Lists, duplicates, and formula setup
- I need to practice list-processing logic (remove duplicates, combine lists, count overlap) and equation order.

6. Graph interpretation and rates
- I need to improve interpreting growth patterns (constant rate vs doubling) and per-user averages from two graphs.

7. Internet concepts
- I need to review core networking ideas like TCP/IP purpose and IPv6 advantages.

8. Efficiency and runtime reasoning
- I need to improve comparing how many times procedures are called and how that changes runtime.

## How I will improve on each topic
I will focus on understanding how and why each answer works, since this is MCQ and not code execution.

- Boolean logic and conditionals
  - I will create mini truth tables and test 2 to 3 sample input cases before choosing an option.

- Algorithm tracing
  - I will trace with a table (variable name, old value, new value, reason) for at least one full loop run.

- Binary and bits
  - I will do short daily drills: 5 binary-decimal conversions, 2 RGB conversions, and 1 bit-capacity question.

- Lists and formulas
  - I will rewrite each formula in words first (for example, “subtract minimum, then divide by remaining count”) before selecting.

- Graph and data analysis
  - I will compare shapes first (same shape vs different shape), then check whether the question is asking for total or average.

- Internet and vocabulary concepts
  - I will make one-page notes with key terms and one example per term (TCP/IP, IPv4 vs IPv6, metadata).

- Runtime and efficiency
  - I will count operations explicitly (calls per loop × number of loops) before deciding on runtime comparisons.

## My study routine before the next MCQ
1. 20 minutes: targeted review of the two weakest topics.
2. 20 minutes: mixed MCQ practice with written elimination notes.
3. 10 minutes: error log update (what I missed, why, what rule fixes it).
4. 10 minutes: quick recap of key rules and vocabulary.

## Strategy I will use during the next MCQ
1. Slow down and annotate each prompt before selecting an answer.
2. Trace at least one concrete example by hand for every logic or algorithm question.
3. Use quick checkpoints for binary/decimal and range conditions.
4. Eliminate at least one wrong choice with evidence before locking my answer.
5. Do a short self-check at the end: “Did I verify all answer choices against the prompt?”

## What strong completion looks like
A strong submission will show that I:
- Finished the MCQ corrections
- Created an issue with clear weakness analysis
- Included a realistic strategy with concrete next steps
- Prepared talking points for the 2-minute individual review

### Quick pattern to improve next time
1. Translate to a tiny test case before choosing.
2. For conditionals, verify all combinations (T/F or edge values).
3. For loops, track variable updates each iteration to avoid skips/infinite loops.
4. For binary/decimal, convert one component at a time and re-check place values.
5. For algorithms, check operation counts (how many calls, how many element accesses).
