## Skill A (Individual): 1-Minute Task Choice
**Chosen task:** Let users upload their own GameBuilder images by adding files to the correct folder and registering them in JSON manifests.

### Problem being solved
Originally, GameBuilder mostly depended on a small built-in set of backgrounds and sprites.
That created a limited-choice problem: users could build game logic, but visual design choices were restricted to whatever assets were preloaded.

### How this task solves it
This task changes asset selection from a fixed list to an extensible list.
Instead of hard-coding every possible image in the main JavaScript, users can add new files and metadata entries, and the builder discovers them dynamically.
That means users are no longer blocked by the original small menu of options; they can use their own art style, theme, and characters.

It also matches the Skill A constraints:
- no new page required
- not a duplicate page task
- reusable by multiple pages/teams, while one person documents it for AP CSP

> After images and JSON are updated, GameBuilder code handles discovery, dropdown population, and runtime use automatically.

### How users upload custom images (in depth workflow)
1. Place image files in the correct folders so the builder can find them by category:
   - Background images go in `images/gamebuilder/bg`
   - Sprite sheets go in `images/gamebuilder/sprites`

2. Register each image in the folder manifest (`index.json`) so the program knows how to use it:
   - `images/gamebuilder/bg/index.json`
   - `images/gamebuilder/sprites/index.json`

3. For backgrounds, include display name, source path, and dimensions (`w`, `h`):
```json
[
  {
    "name": "Forest Night",
    "src": "/images/gamebuilder/bg/forest_night.png",
    "w": 1280,
    "h": 720
  }
]
```

4. For sprites, include source plus animation layout metadata (`rows`, `cols`) so frame slicing works correctly:
```json
[
  {
    "name": "Robot Hero",
    "src": "/images/gamebuilder/sprites/robot_hero.png",
    "w": 384,
    "h": 512,
    "rows": 4,
    "cols": 3
  }
]
```

5. Click **Refresh Assets** in GameBuilder.
   - The asset scan logic reads these manifests, updates internal `assets.bg` / `assets.sprites`, and repopulates dropdown options.
   - If width/height are missing, the loader can infer dimensions from the image file.

6. Select the newly added assets and build as usual.
   - No extra engine rewrite is needed; generated code uses the selected asset path and metadata automatically.

### Why this removes the original limited-choice issue
Before this workflow, users were limited to a small predefined list.
Now the list is open-ended: any valid image can become a selectable background or sprite by adding the file + JSON entry.
So the system moves from "choose from what developers pre-shipped" to "choose from any asset the user provides."

### Algorithm with sequencing, selection, and iteration (inside procedure)
```javascript
function barriers_generate(walls, drawShapes, options = {}) {
    const defs = [];
    const classes = [];
    const visible = options.visible !== undefined ? options.visible : true;

    walls.forEach((w, idx) => {
        const bData = barrier_extract(w, 'wall', idx, { visible: visible });
        const barrierCode = barrier_code(bData);
        defs.push(barrierCode.def);
        classes.push(barrierCode.classEntry);
    });

    const drawnBarriers = (drawShapes || []).filter(s => s.type === 'barrier');
    drawnBarriers.forEach((b, bIdx) => {
        const bData = barrier_extract(b, 'drawn', bIdx, { scaleX: scaleX, scaleY: scaleY });
        const barrierCode = barrier_code(bData);
        defs.push(barrierCode.def);
        classes.push(barrierCode.classEntry);
    });

    return { defs, classes };
}
```
This procedure contains a full algorithm, which means a step-by-step method to solve a problem.
Sequencing: the code runs in order—initialize lists, process wall data, process drawn barriers, then return results.
Selection: the code makes decisions using conditions, such as choosing a default `visible` value and filtering only shapes where `type === 'barrier'`.
Iteration: the code repeats actions over collections using `forEach` loops for walls and drawn barriers.
Together, these parts transform raw user-created data into structured barrier objects the game engine can use.

### Output based on input
```javascript
function renderDrawShapes() {
    ui.drawOverlay.innerHTML = '';
    ui.drawShapes.forEach(shape => {
        const el = document.createElement('div');
        el.className = `draw-rect ${shape.type}`;
        el.style.left = shape.x + 'px';
        el.style.top = shape.y + 'px';
        el.style.width = Math.max(0, shape.width) + 'px';
        el.style.height = Math.max(0, shape.height) + 'px';
        frag.appendChild(el);
    });
    ui.drawOverlay.appendChild(frag);
}
```
```javascript
if (ui.codePlayBtn) ui.codePlayBtn.addEventListener('click', runInRunner);
```
This is visual output because the program draws and updates what the user sees on screen.
The user inputs wall and shape data, and `renderDrawShapes()` turns that data into visible rectangles in the overlay.
When the user presses Run, `runInRunner` uses current settings/code to update the game display.
So input changes (clicks, sliders, drawing) directly control output changes (what appears in the game viewer and overlay).
