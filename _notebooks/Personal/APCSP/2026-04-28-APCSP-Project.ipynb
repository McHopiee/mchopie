## Procedure (with algorithm)
Capture and paste two program code segments you developed during the administration of this task that contain a student-developed procedure that implements an algorithm used in your program and a call to that procedure. 

#### The first program code segment must be a student-developed procedure that: 
- Defines that procedure's name and return type (if necessary)
- Contains and uses one or more parameters that have an effect on the functionality of the procedure 
- Implements an algorithm that includes sequencing, selection, and iteration 

```python
def searchPets(speciesChoice):
    global matchCount
    matchCount = 0
    
    for index in range(len(petNames)):
        result = checkPet(index, speciesChoice)
        if result != "":
            print(result)
    
    if matchCount == 0:
        print("No pets found for that species.")
    else:
        print("Total matches: " + str(matchCount))
```

#### The second program code segment must show where your student-developed procedure in being called in your program. 

```python
userChoice = input("Enter a species to search for: ")
searchPets(userChoice)
```

## List 
Capture and paste two program code segments you developed during the administration of this task that contain a list (or other collection type) being used to manage complexity in your program. 

#### The first program code segment must show how data have been stored in the list 
```python
petNames = ["Bella", "Elliot", "Mango", "Rocky", "Mochi", "Fred", "George", "Miso"]
petSpecies = ["Dog", "Cat", "Parrot", "Dog", "Hamster", "Dog", "Dog", "Cat"]
petAdopted = [True, False, True, False, False, True, True, False]
```

#### The second program code segment must show the data in the same list being used, such as creating a new data from the existing data or accessing multiple elements in the list, as a part of fulfilling the program's purpose 

```python
def checkPet(index, speciesChoice):
    global matchCount
    
    if petSpecies[index].lower() == speciesChoice.lower():
        matchCount += 1
        
        if petAdopted[index]:
            return petNames[index] + " is Adopted"
        else:
            return petNames[index] + " is Available"
    else:
        return ""
```
