```python
petNames = ["Bella", "Whisker", "Mango", "Rocky"]
petSpecies = ["Dog", "Cat", "Parrot", "Dog"]
petAdopted = [True, False, True, False]

matchCount = 0

def checkPet(index, speciesChoice):
    global matchCount
    
    if petSpecies[index].lower() == speciesChoice.lower():
        matchCount = matchCount + 1
        
        if petAdopted[index]:
            return petNames[index] + " is Adopted"
        else:
            return petNames[index] + " is Available"
    else:
        return ""

userChoice = input("Enter a species to search for: ")

for i in range(len(petNames)):
    result = checkPet(i, userChoice)
    if result != "":
        print(result)

if matchCount == 0:
    print("No pets found for that species.")
else:
    print("Total matches: " + str(matchCount))
```

## Row 1: Purpose and Function  
**Purpose:** Help users find pets of a specific species and see whether they are adopted or available  

**Function:** Asks the user for a species, checks each pet in the list, displays matching pets, and shows the total number of matches or a message if none are found  

**Input:** User’s typed species (e.g., "Dog", "Cat", "Fish")  

**Output:** Printed pet names with adoption status and either a total count or a "No pets found" message  

---

## Row 2: Data Abstraction  
**List used:** `petNames`, `petSpecies`, `petAdopted`  

**Data contained:**  
- Pet names (strings)  
- Pet species (strings)  
- Adoption status (booleans)  

**How it manages complexity:** Stores all pet data in organized, indexable lists so the program can access and process multiple pets using the same index  

---

## Row 3: Managing Complexity  
**Why lists are needed:** Without lists, the program would require separate variables like `pet1Name`, `pet2Name`, `pet3Name`, etc. This would make it impossible to use a loop and would require repeating code for every pet. Lists allow the program to efficiently process all pets using iteration.  

---

## Row 4: Procedural Abstraction  
**Procedure:** `checkPet`  

**Parameters:** `index`, `speciesChoice`  

**Called at:** Inside the loop with `result = checkPet(i, userChoice)`  

**What it does:** Checks if a pet matches the selected species, determines adoption status, updates the match count, and returns a formatted result  

---

## Row 5: Algorithm Implementation  
**Sequencing:** Get user input → loop through pets → call procedure → print results → display total or no results message  

**Selection:** `if` statements check whether the species matches and whether the pet is adopted  

**Iteration:** `for` loop iterates through all pets in the lists  

---

## Row 6: Testing  
**Call 1:** `checkPet(0, "Dog")`  
- Tests matching species  
- Returns pet info and increases match count  

**Call 2:** `checkPet(0, "Fish")`  
- Tests non-matching species  
- Returns empty result and does not increase count
